```python
df["daqiqa"]=pd.to_timedelta(df["Duration"]).dt.total_seconds()/60
df["daqiqa"]
```

```python
sns.heatmap(df[["Price","daqiqa"]].corr(), annot=True)
plt.show()
```

```python
stops={
    "non-stop":0,
    "1 stop":1,
    "2 stops":2,
    "3 stops":3,
    "4 stops":4
}
df["Total_stops"]=df["Total_Stops"].replace(stops)
df
```

```python
sns.heatmap(df[["Price","Total_stops"]].corr(), annot=True)
plt.show()
```

```python
sns.heatmap(df[["Price","Total_stops", "daqiqa"]].corr(), annot=True)
plt.show()
```
