```python
df["COMPL_RPY_5YR_RT_NUMERIC"] = pd.to_numeric(df["COMPL_RPY_5YR_RT"], errors='coerce')
df.groupby(by=['REGION', 'Ownership', "Degree_Type"])['COMPL_RPY_5YR_RT_NUMERIC'].agg(['count', 'mean', 'median'])
```

```python
group_cols = ['REGION', 'Ownership', "Degree_Type", "Locale_Short"]
df['COMPL_RPY_5YR_RT_NUMERIC_FILLNA'] = (
    df
    .groupby(by=group_cols)['COMPL_RPY_5YR_RT_NUMERIC']
    .transform(lambda x: x.fillna(x.median()))
    )

df['COMPL_RPY_5YR_RT_NUMERIC_FILLNA_MEAN'] = (
    df
    .groupby(by=group_cols)['COMPL_RPY_5YR_RT_NUMERIC']
    .transform(lambda x: x.fillna(x.mean()))
    )
df
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

df['COMPL_RPY_5YR_RT_NUMERIC'].plot(kind='hist', ax=ax, alpha=0.5, color='blue', bins=20, label='Numeric')
df['COMPL_RPY_5YR_RT_NUMERIC_FILLNA'].plot(kind='hist', ax=ax, alpha=0.5, color='green', bins=20, label='Fillna_Median')
df['COMPL_RPY_5YR_RT_NUMERIC_FILLNA_MEAN'].plot(kind='hist', ax=ax, alpha=0.5, color='red', bins=20, label='Fillna_Mean')

ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.legend()
plt.grid()

plt.title('Histograms of NaN, Fillna_Median, Fillna_Mean')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
df['COMPL_RPY_5YR_RT_NUMERIC'].plot(kind='kde', ax=ax, alpha=0.5, color='blue',  label='Numeric')
df['COMPL_RPY_5YR_RT_NUMERIC_FILLNA'].plot(kind='kde', ax=ax, alpha=0.5, color='green',  label='Fillna_Median')
df['COMPL_RPY_5YR_RT_NUMERIC_FILLNA_MEAN'].plot(kind='kde', ax=ax, alpha=0.5, color='red',  label='Fillna_Mean')
ax.set_xlabel('Value')
ax.set_ylabel('Frequency')
ax.legend()
plt.grid()
plt.title('Histograms of NaN, Fillna_Median, Fillna_Mean')
plt.show()
```
